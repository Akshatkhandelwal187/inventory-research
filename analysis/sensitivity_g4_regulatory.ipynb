# Sensitivity G4 — regulatory comparison under non-stationary demand

Two parallel sweeps (30 points each) at the shared reference parameters:
- **Sweep A:** vary `p_c ∈ [0, 5]`, `C_cap` fixed at BASE.
- **Sweep B:** vary `C_cap ∈ [0, 30]`, `p_c` fixed at BASE.

All three regimes (tax / cap-and-trade / strict cap) are evaluated per grid point.

**Outputs (regenerated by this notebook):**
- `analysis/sensitivity_g4_by_pc.csv` and `analysis/sensitivity_g4_by_ccap.csv`
- `analysis/figures/sensitivity_g4.pdf` — publication 2×2 figure
- `analysis/sensitivity_g4_findings.md` — takeaway

## 1. Run the sweep

```python
import os, sys
from pathlib import Path

ROOT = Path.cwd()
if ROOT.name == "analysis":
    ROOT = ROOT.parent
    os.chdir(ROOT)
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

from analysis.sweep_g4 import run, write_findings
csv_a, csv_b = run()
md_path = write_findings()
print("csv A:", csv_a.relative_to(ROOT))
print("csv B:", csv_b.relative_to(ROOT))
print("pdf  :", (ROOT / "analysis" / "figures" / "sensitivity_g4.pdf").relative_to(ROOT))
print("md   :", md_path.relative_to(ROOT))
```

## 2. Raw data

```python
import pandas as pd
df_a = pd.read_csv("analysis/sensitivity_g4_by_pc.csv")
df_b = pd.read_csv("analysis/sensitivity_g4_by_ccap.csv")
print("sweep A shape:", df_a.shape, " sweep B shape:", df_b.shape)
df_a.head(6)
```

## 3. Inline preview

Cost vs `p_c` (sweep A) and cost vs `C_cap` (sweep B) for all three regimes. Full publication 2×2 figure at `analysis/figures/sensitivity_g4.pdf`.

```python
%matplotlib inline
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(9, 3.4))
for policy, sub in df_a.groupby("policy"):
    axes[0].plot(sub["p_c"], sub["cost"], label=policy.replace("_", " "))
axes[0].set_xlabel("carbon price  p_c"); axes[0].set_ylabel("total cost")
axes[0].set_title("sweep A: cost vs p_c"); axes[0].legend()
for policy, sub in df_b.groupby("policy"):
    axes[1].plot(sub["C_cap"], sub["cost"], label=policy.replace("_", " "))
axes[1].set_xlabel("emissions cap  C_cap"); axes[1].set_ylabel("total cost")
axes[1].set_title("sweep B: cost vs C_cap")
fig.tight_layout(); plt.show()
```
