## 3. Inline preview

Two-panel preview from the CSV. The full publication figure (4 panels) is at `analysis/figures/sensitivity_g1.pdf`.

```python
%matplotlib inline
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(9, 3.2), sharex=True)
for policy, sub in df.groupby("policy"):
    axes[0].plot(sub["n"], sub["T_star"], label=policy.replace("_", " "))
    axes[1].plot(sub["n"], sub["cost"],   label=policy.replace("_", " "))
axes[0].set_ylabel("T*");      axes[0].set_xlabel("power-demand exponent n")
axes[1].set_ylabel("total cost"); axes[1].set_xlabel("power-demand exponent n")
axes[0].legend(); fig.suptitle("G1 preview — T* and cost vs n"); fig.tight_layout()
plt.show()
```
