# Sensitivity G5 — demand-investment coupling under power demand

Coupling coefficient `D_1` varies across `[0, 6.0]` (30 points) under two regimes:
- cap-and-trade at BASE,
- strict cap at a tight `C_cap = 5`.

Reveals the corner transition where stronger coupling drives `G* → 0` under cap-and-trade, while strict cap forces continued investment.

## 3. Inline preview

`G*` against coupling `D_1` for both regimes — shows the cap-and-trade corner transition. Full publication 4-panel figure at `analysis/figures/sensitivity_g5.pdf`.

```python
%matplotlib inline
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(6, 3.6))
for regime, sub in df.groupby("regime"):
    ax.plot(sub["D1"], sub["G_star"], label=regime.replace("_", " "))
ax.set_xlabel("demand-coupling coefficient  D_1")
ax.set_ylabel("G* (per-time investment)")
ax.legend(); ax.set_title("G5 preview — G* vs D_1 across regimes")
fig.tight_layout(); plt.show()
```
