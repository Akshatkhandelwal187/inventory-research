## 3. Inline preview

`G*` heatmap with the analytic boundary `p_c · a = 1` overlaid (corner regime below, interior above). Full publication 2-panel figure at `analysis/figures/sensitivity_g3.pdf`.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

G_arr = df.pivot(index="a", columns="p_c", values="G_star").values
p_c = sorted(df["p_c"].unique()); a = sorted(df["a"].unique())
extent = [p_c[0], p_c[-1], a[0], a[-1]]

fig, ax = plt.subplots(figsize=(6, 4))
im = ax.imshow(G_arr, origin="lower", aspect="auto", extent=extent, cmap="viridis")
pc_dense = np.linspace(max(p_c[0], 1e-6), p_c[-1], 200)
ab = 1.0 / pc_dense
mask = (ab >= a[0]) & (ab <= a[-1])
ax.plot(pc_dense[mask], ab[mask], color="white", linestyle=":", linewidth=1.4, label="p_c · a = 1")
ax.set_xlabel("carbon price  p_c"); ax.set_ylabel("green-tech efficiency  a")
ax.set_title("G3 preview — G* heatmap with analytic boundary")
ax.legend(loc="upper right", labelcolor="white", facecolor="black", framealpha=0.4)
fig.colorbar(im, ax=ax, shrink=0.85)
fig.tight_layout(); plt.show()
```
